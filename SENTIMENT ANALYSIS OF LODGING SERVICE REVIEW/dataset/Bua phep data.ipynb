# check other_sentiment ratio for each aspects

```python
query_result = {}
for item in aspects:
  print(item)
  cols = list(aspect_df[item].columns)[1:]
  cols.remove(item)
  print(cols)

  query_result[item] = pd.DataFrame(aspect_df[item].value_counts(subset=cols).reset_index())
  
  query_result[item].rename(columns = {0:'count'}, inplace = True)
  
  query_result[item]['review_has_other_sentiments'] = (query_result[item][cols]>0).any(axis="columns")
```

```python
for item in aspects:
  print('{aspect} has {count} reviews'.format(aspect = item, count = aspect_df[item].Review.count()))
  print('which consists of the follow ',query_result[item].groupby(by='review_has_other_sentiments').sum()['count'])
```

```python
crawl_list = os.listdir('data_crawl')
crawl_list
```

```python
len(crawl_list)
```

```python
%pwd
```

```python
crawl_df = {}
```

```python
for item in crawl_list:
  dot = '.'
  tmp = item.split(sep = dot, maxsplit = 1)
  dfname = tmp[0]
  ext = tmp[1]
  path = 'data_crawl/' + item
  print(tmp)
  if (ext == 'csv'):
    crawl_df[dfname] = pd.read_csv(path, header = None)
  elif (ext == 'xlsx'):
    crawl_df[dfname] = pd.read_excel(path, header = None)
```

```python
crawl_df['data_khach_san']
```

```python
crawl_df['data_giai_tri'].drop(0, axis = 1, inplace = True)
crawl_df['data_giai_tri'].drop(0, axis = 0, inplace = True)
crawl_df['data_giai_tri']
```

```python
crawl_df['data_khach_san'].drop(0, axis = 1, inplace = True)
crawl_df['data_khach_san'].drop(0, axis = 0, inplace = True)
crawl_df['data_khach_san']
```

```python
crawl_df['data_nha_hang_tripadvisor'].drop(0, axis = 0, inplace = True)
crawl_df['data_nha_hang_tripadvisor']
```

```python
dfnames = list(crawl_df.keys())
for name in dfnames:
  crawl_df[name].set_axis(['Review'], axis = 1, inplace = True)
  crawl_df[name].reset_index(drop=True, inplace=True)
```

```python
crawl_df
```

```python
crawl_df['data_khach_san'].drop_duplicates(inplace=True)
```

```python
crawl_df['data_khach_san'].reset_index(drop=True, inplace=True)
```

```python
import zipfile

crawl_df['data_khach_san'].to_csv('data_khach_san_crawl.csv',index=None,sep=",",header=True,encoding='utf-8-sig')

with zipfile.ZipFile('data_khach_san_crawl.zip', 'w') as zf:
    zf.write('data_khach_san_crawl.csv')
```
